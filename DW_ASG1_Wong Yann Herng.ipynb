```python
def find_boundaries(df, variable):

    # distance passed as an argument, gives us the option to
    # estimate 1.5 times or 3 times the IQR to calculate
    # the boundaries.

    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * 1.5)
    upper_boundary = df[variable].quantile(0.75) + (IQR * 1.5)

    return upper_boundary, lower_boundary

```

```python
def diagnostic_plots(df, variable):
    # function takes a dataframe (df) and
    # the variable of interest as arguments

    # define figure size
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30, kde=True)  
    plt.title('Histogram')

    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('popularity quantiles')

    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
def find_frequent_labels(df, variable, tolerance):
    
    temp = df[variable].value_counts()  / len(df)
    
    non_rare = [x for x in temp.loc[temp>tolerance].index.values]
    
    return non_rare
```

In order to evaluate the model performance of a linear regression model, a range of measures can be used, but in this ASG1 we will focus on only 2:

### Feature Engineering 
For my feature engineering, I have chosen robust scaling as my scaling method. This is because robust scaling is useful when it comes to dealing with outliers. It removes the median and scales the data according to the interquartile range (IQR). Thus this would reduce the effect of the extreme values on the scaling process. 

Comparing Robust scaling to other scaling methods, like standardization, standardization could only be applied to normal distribution for effective scaling, but this scaling method is sensitive to outliers which could skew the mean and standard deviation, leading to less effective scaling for non-normally distributed data. For Mean normalization, it would be more useful for data centering around 0 however it is also sensitive to outliers just like standardization, which could distort the mean and range. For minmax scaling, it is applicable when it comes to maintaining relationships between data points, but it is extremely sensitive to outliers as a single outlier can affect the scale. Lastly, for maximum absolute scaling, it is useful when the data is already centered around 0. On the other hand, it is less effective when it comes to handling outliers as the scaling is easily affected by the maximum values. Since robust scaling is the only scaling method which could deal with outliers, I have chosen the robust scaling method to scale the data since the dataset contains several outliers like the Instrumentalness variable, and it has the ability to provide a more consistent scaling across features.
